# Data Cleaning

```python
# Remove Exact Duplicate Rows

print(f"Shape Before : {df.shape}")

duplicate_count = df.duplicated().sum()
print(f"Duplicate Rows : {duplicate_count}")

# Remove duplicates
df = df.drop_duplicates().reset_index(drop=True)

print(f"\nShape After : {df.shape}")
```

```python
# Clean AskPrice (Target Variable)

# Remove currency symbol, commas, and spaces
df["AskPrice"] = (
    df["AskPrice"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

# Convert to numeric
df["AskPrice"] = pd.to_numeric(df["AskPrice"], errors="coerce")

# Check results
print(df["AskPrice"].head())

print("\nData Type:")
print(df["AskPrice"].dtype)

print("\nMissing Values:")
print(df["AskPrice"].isnull().sum())
```

```python
# Clean kmDriven

# Remove "km", commas, and spaces
df["kmDriven"] = (
    df["kmDriven"]
    .str.replace("km", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)

# Convert to numeric
df["kmDriven"] = pd.to_numeric(df["kmDriven"], errors="coerce")

# Check results
print(df["kmDriven"].head())

print("\nData Type:")
print(df["kmDriven"].dtype)

print("\nMissing Values:")
print(df["kmDriven"].isnull().sum())
```

```python
# Convert PostedDate to Datetime

df["PostedDate"] = pd.to_datetime(
    df["PostedDate"],
    format="%b-%y"
)

print(df["PostedDate"].head())

print("\nData Type:")
print(df["PostedDate"].dtype)
```

```python
# Standardize FuelType

df["FuelType"] = df["FuelType"].replace({
    "hybrid": "Hybrid"
})

print(df["FuelType"].value_counts())
```

```python
# Check Cars with Very High Age

df[df["Age"] > 30][["Brand", "model", "Year", "Age", "AskPrice"]]
```

```python
df[df["Year"] < 1950][
    ["Brand", "model", "Year", "Age", "AskPrice"]
]
```

```python
# Validate Age

current_year = 2024

age_check = df["Age"] == (current_year - df["Year"])

print("All Ages Correct:", age_check.all())

print("\nIncorrect Rows:")
print(df.loc[~age_check, ["Year", "Age"]].head())
```

```python
# Remove Impossible Manufacturing Years

print("Shape Before:", df.shape)

df = df[df["Year"] >= 1950].reset_index(drop=True)

print("Shape After:", df.shape)
```

```python
# Final Data Quality Check

print("Dataset Shape:")
print(df.shape)

print("\n")


print("Missing Values:")
print("-"*20)
print(df.isnull().sum())

print("\n")


print("Duplicate Rows:")
print("-"*20)
print(df.duplicated().sum())

print("\n")


print("Data Types:")
print("-"*20)
print(df.dtypes)
```

```python
# Inspect Remaining Duplicate Rows

remaining_duplicates = df[df.duplicated(keep=False)]

print("Number of duplicate rows:", remaining_duplicates.shape[0])

remaining_duplicates.sort_values(by=list(df.columns)).head(20)
```

```python
# Remove Remaining Duplicate Rows

print("Shape Before:", df.shape)
print("Duplicate Rows:", df.duplicated().sum())

df = df.drop_duplicates().reset_index(drop=True)

print("\nShape After:", df.shape)
print("Duplicate Rows:", df.duplicated().sum())
```

```python
# Feature Extraction from PostedDate

df["Posted_Year"] = df["PostedDate"].dt.year
df["Posted_Month"] = df["PostedDate"].dt.month
df["Posted_Month_Name"] = df["PostedDate"].dt.month_name()

df[["PostedDate", "Posted_Year", "Posted_Month", "Posted_Month_Name"]].head()
```

```python
# Save the Cleaned Dataset

df.to_csv("used_cars_clean.csv", index=False)
print("Cleaned Dataset Saved Successfully!")
```
