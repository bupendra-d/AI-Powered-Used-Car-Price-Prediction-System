```python
import pandas as pd
import json
```

```python
df = pd.read_csv("feature_engineered_dataset.csv")
df.head()
```

```python
brand_model = (
    df.groupby("Brand")["model"]
      .unique()
      .apply(lambda x: sorted(list(x)))
      .sort_index()
      .to_dict()
)

with open("brand_model.json", "w") as f:
    json.dump(brand_model, f, indent=4)

print("brand_model.json created successfully!")
```
